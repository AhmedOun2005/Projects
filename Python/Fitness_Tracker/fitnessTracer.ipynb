### Drop Null Values

```python
df.drop('Reviews',axis='columns',inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['Rating (Out of 5)']=df['Rating (Out of 5)'].fillna(df['Rating (Out of 5)'].median())
df.isnull().sum()
```

### change Datatype for Prices

```python
df['Selling Price']=df['Selling Price'].str.replace(',','')
df['Original Price']=df['Original Price'].str.replace(',','')
df['Selling Price']=df['Selling Price'].astype(float)
df['Original Price']=df['Original Price'].astype(float)
df.info()
```

### Drop Duplicated Values

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

### Transforming some Error values in Color

```python
df['Color']=df['Color'].str.replace(' ','')
df['Color']=df['Color'].str.title()
df['Color']=df['Color'].str.replace('Grey','Gray')
```

```python
df
```

```python
df.groupby('Color').agg({'Color':'count'})
```

```python
df['Color']=df['Color'].apply(lambda x : str(sorted(str(x).split(','))))
```

```python
df.groupby('Color').agg({'Color':'count'})
```

```python
df
```

### Add column for Discount

```python
df['Discount']=df['Original Price']-df['Selling Price']
df
```
